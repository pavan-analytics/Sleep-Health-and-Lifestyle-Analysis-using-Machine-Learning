```python
df['Sleep Disorder']=df['Sleep Disorder'].str.replace('No Disorder','0')
df['Sleep Disorder']=df['Sleep Disorder'].str.replace('Sleep Apnea','1')
df['Sleep Disorder']=df['Sleep Disorder'].str.replace('Insomnia','2')
df['BMI Category']=df['BMI Category'].str.replace('Normal weight','Normal')
df['Sleep Disorder']=df['Sleep Disorder'].astype(int)
```

```python
X=df[["Sleep Duration","Gender","BMI Category","Diastolic BloodPressure","Age","Physical Activity Level"]]
y=df["Sleep Disorder"]
print(X.shape)
print(y.shape)
```

```python
# let us split the data as train and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train.dtypes
```

```python
X_train_cat=X_train.select_dtypes(include=['object'])
X_train_num=X_train.select_dtypes(include=['float64','int64'])

print(X_train_cat.dtypes)
print(X_train_num.dtypes)
```

```python
X_test_cat=X_test.select_dtypes(include=['object'])
X_test_num=X_test.select_dtypes(include=['float64','int64'])

print(X_test_cat.dtypes)
print(X_test_num.dtypes)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_num=pd.DataFrame(sc.fit_transform(X_train_num),
                         columns=sc.get_feature_names_out(),
                         index=X_train_num.index)
X_test_num=pd.DataFrame(sc.transform(X_test_num),
                        columns=sc.get_feature_names_out(),
                        index=X_test_num.index)
print(X_train_num.shape)
print(X_test_num.shape)
```

```python
#feature scaaling
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
X_train_cat=pd.DataFrame(oe.fit_transform(X_train_cat),
                         columns=oe.get_feature_names_out(),
                         index=X_train_cat.index)
X_test_cat=pd.DataFrame(oe.transform(X_test_cat),
                        columns=oe.get_feature_names_out(),
                        index=X_test_cat.index)
print(X_train_cat.shape)
print(X_test_cat.shape)
```

```python
#concatenation
X_train_trans=pd.concat([X_train_num,X_train_cat],axis=1)
X_test_trans=pd.concat([X_test_num,X_test_cat],axis=1)
print(X_train_trans.shape)
print(X_test_trans.shape)
```

Decision Tree Classification

Hyperparamter tuning for Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

n_folds=5
dt=DecisionTreeClassifier()
hyper_parameters=[{"max_depth":range(2,20,5)}]

dt_model=GridSearchCV(estimator=dt,
                      param_grid=hyper_parameters,
                      cv=n_folds,
                      return_train_score=True,
                      verbose=1)
dt_model.fit(X_train_trans,y_train)
```

```python
print("Best Parameters sets found on train set")
print(dt_model.best_params_)
print(dt_model.best_estimator_)
print()

print("Score of Train set: ",dt_model.best_score_)
print("Score of test set: ",dt_model.score(X_test_trans,y_test))
```

```python
cv_results=pd.DataFrame(dt_model.cv_results_)
cv_results.head()
```

```python
plt.plot(cv_results['param_max_depth'],cv_results['mean_train_score'])
plt.plot(cv_results['param_max_depth'],cv_results['mean_test_score'])
plt.xlabel("max_depth")
plt.ylabel("Accuracy")
plt.legend(['train_accuracy','test_accuracy'], loc='upper left')
plt.show()
```

RandomForest Classifier

Building the Model - Stacking

```python
from sklearn.ensemble import StackingClassifier

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

estimators = [
    ('lr', LogisticRegression()),
    ("knn",KNeighborsClassifier(10)),
    ('dt',DecisionTreeClassifier(random_state=42))
]

stack_classifier = StackingClassifier(estimators=estimators,
                                      final_estimator=DecisionTreeClassifier())

stack_classifier.fit(X_train_trans, y_train)
```

```python
y_pred=stack_classifier.predict(X_test_trans)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

print("*"*10, "Accuracy","*"*10)
print(accuracy_score(y_test,y_pred))

print("*"*10, "confusion matric","*"*10)
print(confusion_matrix(y_test,y_pred))

print("*"*10, "classification report","*"*10)
print(classification_report(y_test,y_pred))
```

HyperParamter for RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

n_folds=5

hyper_parameters=[{"max_depth":range(2,20,5)}]

rf=RandomForestClassifier()
# fit tree on training data
rf_model=GridSearchCV(estimator=rf,
                             param_grid=hyper_parameters,
                             cv=n_folds,
                             scoring="accuracy",
                             return_train_score=True,
                             verbose=1)
rf_model.fit(X_train_trans, y_train)
```

```python
print("Best parameters set found on train set")
print(rf_model.best_params_)
print(rf_model.best_estimator_)
print()

print('Score on Test Data: ', rf_model.score(X_test_trans, y_test))
```
